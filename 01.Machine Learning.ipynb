## Aprendizaje automático (Machine learning)
Cuando la mayoría de las personas escuchan el térmido aprendizaje de máquina o autónomo, una de las primeras imágenes que viene a la mente un robot T-800 con la capacidad de aprender desde un saludo hasta aniquilar a la raza humana de la faz de la tierra.
![ml-boot](https://www.atriainnovation.com/wp-content/uploads/2021/02/portada-1080x675.jpg)
El aprendizaje automático es la ciencia y tecnología de dotar a sistemas de computo con algoritmos (sin ser explícitamente programados) capaces de decifrar patrones que no son tan aparentes a partir del análisis de una serie de datos. 

### Aprendizaje automático supervisado
El algoritmo es entrenado con diferentes descriptores (predictors) relacionados con sus respectivas etiquetas de clasificación, creando un modelo que aprendió de manera supervisada o guiada. 
![sup-ml](https://cdn-images-1.medium.com/max/1600/1*Iz7bCLrPTImnBDOOEyE3LA.png)
Lista de los algoritmos de aprendizaje atuomático supervisado más utilizados
* Regresión lineal
* Regresión logística
* k-nearest neighbors
* Suppor vector machine (SVM)
* Decision Tree y Random Forest 


### Aprendizaje automático no supervisado
El modelo es entrenado sin la posibilidad de conocer la posible respuesta, intenta entonces separar o agrupar los datos, basados en sus descriptores
![usup-ml](https://miro.medium.com/max/1204/0*Lo2t0XnKPB2JN-_t)
Lista de los algoritmos de aprendizaje atuomático sin supervisión más utilizados
* K-menas
* Clustering 
* Principal Component Analysis (PCA)
* Kernel PCA

### Aprendizaje automático reforzado
Bajo el contexto del aprendizaje reforzado, el modelo (llamado agente) interactua con el ambiente y decide las tareas a cumplir, como producto de esta acción recibe recompensas o penalidades
