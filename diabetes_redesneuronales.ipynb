# Diabetes gestacional con redes neuronales

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/jrmaza/datasets/main/pima-indians-diabetes.csv'
diabetes = pd.read_csv(url)
diabetes
```

```python
diabetes.columns = ['preg','glu','pre','ski','desc1','desc2','desc3','age','diagnosis']
diabetes
x = diabetes.drop(['diagnosis'], axis=1)
x
```

```python
y = diabetes.drop(['preg', 	'glu', 	'pre', 	'ski', 	'desc1', 	'desc2', 	'desc3', 	'age'], axis=1)
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

```python
modelo_diabetes = Sequential()
modelo_diabetes.add(Dense(12, input_shape=(8,), activation='relu'))
modelo_diabetes.add(Dense(8, activation='relu'))
#modelo_diabetes.add(Dense(4, activation='relu'))
modelo_diabetes.add(Dense(1, activation='sigmoid')) # -> sigmoid los datos son binarios 0 y 1

#12-8-1
```

```python
modelo_diabetes.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
resultado = modelo_diabetes.fit(x,y, epochs=500, batch_size=10)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pÃ©rdida")
plt.plot(resultado.history["loss"])
```
